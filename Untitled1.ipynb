```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
Dynamic Top-N Band Selection from STFT (for Pink Noise Masking)
- Load WAV
- STFT (25 ms window, 10 ms hop @16kHz)
- Aggregate energy into mel bands (10 bands from 125 Hz to 8 kHz)
- Per-frame dynamic Top-N selection based on band energy or SNR
- Visualize spectrogram and selected bands over time

Requirements: numpy, librosa, matplotlib
Usage:
  1) Edit `audio_file` path below to your WAV file
  2) Run: python dynamic_topN_mask.py
"""

import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt

# --- User: set your WAV path here ---
audio_file = "C:\\Users\\ipl1\\Desktop\\IPL\\세미나(25하계)\\data\\sound\\audio\\audio\\16000\\1-137-A-32.wav"

# --- Parameters ---
sr = 16000
win_ms = 25
hop_ms = 10
n_fft = 512
win_length = int(sr * win_ms / 1000.0)  # 25 ms -> 400 samples
hop_length = int(sr * hop_ms / 1000.0)  # 10 ms -> 160 samples

# Band aggregation (mel is simple and effective; you can swap to 1/3-octave if you prefer)
n_bands = 10
fmin, fmax = 125.0, 8000.0

# Dynamic selection
top_k = 3         # pick the top-K bands per frame
use_snr = True    # if False, uses absolute band energy
noise_percentile = 10.0  # percentile for baseline/noise floor estimation (per band)

# Soft mask params (optional; used for plotting suggested weights)
T_low = 0.0       # dB SNR threshold to start ramp
T_high = 6.0      # dB SNR where mask saturates
eps = 1e-10

# --- Load audio ---
y, _ = librosa.load(audio_file, sr=sr, mono=True)

# --- STFT ---
S = librosa.stft(
    y,
    n_fft=n_fft,
    hop_length=hop_length,
    win_length=win_length,
    window="hann",
    center=True,
)
S_mag = np.abs(S)
S_pow = S_mag**2  # power spectrogram
S_db = librosa.amplitude_to_db(S_mag, ref=np.max)

# --- Band aggregation (mel filterbank) ---
mel_fb = librosa.filters.mel(
    sr=sr,
    n_fft=n_fft,
    n_mels=n_bands,
    fmin=fmin,
    fmax=fmax,
    htk=True,
    norm=None,
)
# Band power over time: (n_bands, n_frames)
band_pow = mel_fb @ S_pow

# --- Baseline/noise floor per band (for SNR) ---
if use_snr:
    noise_floor = np.percentile(band_pow, noise_percentile, axis=1, keepdims=True) + eps
    snr = 10.0 * np.log10((band_pow + eps) / noise_floor)
    score = snr  # higher is more noisy relative to baseline
else:
    score = 10.0 * np.log10(band_pow + eps)  # dB energy

# --- Dynamic Top-N selection per frame ---
n_frames = score.shape[1]
sel_mask = np.zeros_like(score, dtype=np.float32)  # (n_bands, n_frames), 1 where selected

for t in range(n_frames):
    col = score[:, t]
    # indices of top-k bands
    top_idx = np.argsort(col)[::-1][:top_k]
    sel_mask[top_idx, t] = 1.0

# --- Optional soft weights based on SNR ramp ---
if use_snr:
    # Normalize SNR into [0,1] between T_low and T_high
    soft = np.clip((snr - T_low) / (T_high - T_low), 0.0, 1.0)
    # Keep weights only on selected bands (others set to 0 for clarity)
    soft_weights = soft * sel_mask
else:
    soft_weights = sel_mask.copy()

# --- Visualization ---
fig, axs = plt.subplots(3, 1, figsize=(12, 10), constrained_layout=True)

# 1) STFT magnitude spectrogram (dB)
img = librosa.display.specshow(
    S_db,
    sr=sr,
    hop_length=hop_length,
    x_axis="time",
    y_axis="hz",
    ax=axs[0],
)
axs[0].set_title("STFT Magnitude (dB)")
fig.colorbar(img, ax=axs[0], format="%+2.0f dB")

# 2) Band energy/SNR over time (as heatmap)
if use_snr:
    img2 = axs[1].imshow(score, aspect="auto", origin="lower",
                         extent=[0, n_frames*hop_ms/1000.0, fmin, fmax])
    axs[1].set_title(f"Band SNR (dB) — {n_bands} mel bands {int(fmin)}–{int(fmax)} Hz")
else:
    img2 = axs[1].imshow(score, aspect="auto", origin="lower",
                         extent=[0, n_frames*hop_ms/1000.0, fmin, fmax])
    axs[1].set_title(f"Band Energy (dB) — {n_bands} mel bands {int(fmin)}–{int(fmax)} Hz")
axs[1].set_ylabel("Frequency (Hz)")
axs[1].set_xlabel("Time (s)")
fig.colorbar(img2, ax=axs[1])

# Tick labels for bands
band_centers = librosa.mel_frequencies(n_mels=n_bands, fmin=fmin, fmax=fmax, htk=True)
axs[1].set_yticks(band_centers)
axs[1].set_yticklabels([f"{int(fc)}" for fc in band_centers])

# 3) Dynamic Top-K selection mask (and optional soft weights)
axs[2].imshow(soft_weights, aspect="auto", origin="lower",
              extent=[0, n_frames*hop_ms/1000.0, fmin, fmax])
axs[2].set_title(f"Selected Bands (Top-{top_k}) — brighter = stronger weight")
axs[2].set_ylabel("Frequency (Hz)")
axs[2].set_xlabel("Time (s)")
axs[2].set_yticks(band_centers)
axs[2].set_yticklabels([f"{int(fc)}" for fc in band_centers])

plt.show()

# --- Print a quick summary ---
active_ratio = sel_mask.mean(axis=1)
print("Per-band selection ratio (fraction of frames selected):")
for i, fc in enumerate(band_centers):
    print(f"  Band {i:02d} (~{int(fc)} Hz): {active_ratio[i]*100:.1f}% frames selected")

print("\nTip: Next step is to map these selected (or weighted) bands to pink-noise gain per band,")
print("     e.g., convert soft_weights to dB gains with a cap (±4~6 dB) and apply smoothing.")
```
