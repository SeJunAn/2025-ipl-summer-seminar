<h2>EDA</h2>

직업 별 취침시간 : 판매 대리인, S/W개발자, 과학자가 낮음 / 엔지니어, 의사, 변호사가 높음
성별 별 취침시간 : 여성이 남성보다 **조금** 더 높음
수면시간과 수면의 질 : 수면시간이 길수록 수면의 질이 증가 -> 선형으로 증가

<h3>직업별 취침시간</h3>

```python
# Occupation 값별 Sleep Duration 평균 계산
mean_sleep = data.groupby('Occupation')['Sleep Duration'].mean().reset_index()

# 시각화
plt.figure(figsize=(8, 4))
sns.barplot(data=mean_sleep, x='Occupation', y='Sleep Duration', palette='pastel')

plt.title('Average Sleep Duration by Occupation')
plt.ylabel('Average Sleep Duration (hours)')
plt.xlabel('Occupation')
plt.ylim(0, mean_sleep['Sleep Duration'].max() + 1)
plt.xticks(rotation=45)
plt.show()
```

<h3>성별별 취침시간</h3>

```python
# Gender 값별 Sleep Duration 평균 계산
mean_sleep = data.groupby('Gender')['Sleep Duration'].mean().reset_index()

# 시각화
plt.figure(figsize=(8, 4))
sns.barplot(data=mean_sleep, x='Gender', y='Sleep Duration', palette='pastel')

plt.title('Average Sleep Duration by Gender')
plt.ylabel('Average Sleep Duration (hours)')
plt.xlabel('Gender')
plt.ylim(0, mean_sleep['Sleep Duration'].max() + 1)
plt.xticks(rotation=45)
plt.show()
```

<h3>수면시간과 수면의 질 상관관계</h3>

```python
corr = data['Sleep Duration'].corr(data['Quality of Sleep'])
print(f"Pearson correlation: {corr:.2f}")
```

```python
plt.scatter(data['Sleep Duration'], data['Quality of Sleep'])
plt.title("Sleep duration - Quality of Sleep")
plt.xlabel("Sleep duration")
plt.ylabel("Quality of Sleep")
plt.plot()
```

<h3>수면의 질과 관련된 상관계수</h3>

```python
num_data = data.select_dtypes(include=['number'])
```

```python
corr_matrix = num_data.corr()
```

```python
# 3. 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numeric Features (Including Quality of Sleep)')
plt.show()
```

- 나이가 많음 -> 수면의 질 증가<br>
- 이거의 근거는? : 나이가 많음 -> 수면시간 증가 -> 수면의 질 증가<br>
- 스트레스는 수면의 질과 수면시간에 영향을 받음
