<h3>직업별 취침시간</h3>

```python
# Occupation 값별 Sleep Duration 평균 계산
mean_sleep = data.groupby('Occupation')['Sleep Duration'].mean().reset_index()

# 시각화
plt.figure(figsize=(8, 4))
sns.barplot(data=mean_sleep, x='Occupation', y='Sleep Duration', palette='pastel')

plt.title('Average Sleep Duration by Occupation')
plt.ylabel('Average Sleep Duration (hours)')
plt.xlabel('Occupation')
plt.ylim(0, mean_sleep['Sleep Duration'].max() + 1)
plt.xticks(rotation=45)
plt.show()
```

<h3>성별별 취침시간</h3>

```python
# Gender 값별 Sleep Duration 평균 계산
mean_sleep = data.groupby('Gender')['Sleep Duration'].mean().reset_index()

# 시각화
plt.figure(figsize=(8, 4))
sns.barplot(data=mean_sleep, x='Gender', y='Sleep Duration', palette='pastel')

plt.title('Average Sleep Duration by Gender')
plt.ylabel('Average Sleep Duration (hours)')
plt.xlabel('Gender')
plt.ylim(0, mean_sleep['Sleep Duration'].max() + 1)
plt.xticks(rotation=45)
plt.show()
```

- 환자ID : 나이(0.98), 수면의 질(0.66), 신체활동 레벨(0.57)
- 나이 : 환자ID(0.98), 수면의 질(0.71), 신체활동 레벨(0.6), 수면시간(0.55)
- 수면시간 : 수면의 질(0.92), 스트레스(-0.85), 나이(0.55)
- 수면의 질 : 수면시간(0.92), 스트레스(-0.87), 나이(0.71), 환자ID(0.66), 심박수(-0.53)
- 신체활동 레벨 : 걸음 수(0.78), 나이(0.6), 환자 ID(0.57)
- 스트레스 : 수면의 질(-0.87), 수면시간(-0.85), 심박수(0.6)
- 심박수 : 스트레스(0.6), 수면의 질(-0.53)
- 걸음 수 : 신체활동 레벨(0.78)

**추론**<br>
- 환자ID는 나이의 영향을 받는다. -> 환자ID는 나이순으로 정렬되어 있기 때문.
- 나이가 높으면 수면의 질이 증가한다.
- 수면시간이 증가하면 수면의 질이 증가한다. / 스트레스는 감소한다. / 나이는 증가
- 수면의 질이 높을수록 수면시간이 증가, 스트레스는 감소

**가설**
- 나이가 증가할수록 수면의 질이 증가한다.
- 수면시간은 수면의 질과 스트레스에 영향을 받는다.
- 수면의 질은 나이, 수면시간, 스트레스에 영향을 받는다.
- 신체활동 레벨은 걸음수에 영향을 받는다

**정리**
- 스트레스↑ → 수면의 질 ↓ → 취침시간 ↓

<h2>다중회귀 분석</h2>

```python
import statsmodels.api as sm
```

```python
# (1) 수치형만 가져오되, 식별자/문자열 등은 제거
num_cols = [
    'Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level',
    'Stress Level', 'Heart Rate', 'Daily Steps'
]
use = data.loc[:, [c for c in num_cols if c in data.columns]]

# (2) 결측치가 있으면 X/y가 서로 다른 행을 갖게 되니, 합쳐서 한 번에 dropna
df = use.dropna().copy()

y = df['Quality of Sleep']  # 종속변수
X = df.drop(columns=['Quality of Sleep'])  # 독립변수에서 반드시 제거
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()  # 대문자 X로 사용
```

```python
print(model.summary())
```
